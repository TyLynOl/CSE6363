# K-Means Clustering

This notebook demonstrates using K-Means clustering on [house data](https://www.kaggle.com/datasets/yasserh/housing-prices-dataset). The first goal is to use K-Means to organize the houses into 3 clusters based on the price.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data
data = pd.read_csv('data/Housing.csv')
data.head()
```

```python
# Select only the price and area columns
X = data[['price', 'area']]

# Standardize the data
X = (X - X.mean()) / X.std()
```

```python
# Perform K-Means with 3 clusters on the price and area columns
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Add the cluster labels to the data
data['cluster'] = kmeans.labels_

# Plot the data
plt.scatter(data['price'], data['area'], c=data['cluster'])
plt.xlabel('Price')
plt.ylabel('Area')
plt.title('K-Means with 3 clusters')
plt.show()
```
